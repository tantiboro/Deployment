```python
#loading the dataset
df = pd.read_csv('/content/drive/MyDrive/output.csv', encoding='utf-8')
```

```python
df.head()
```

```python
#define figure size
ax = sns.set(rc={'figure.figsize':(15,8)})
#plot the countplot
ax = sns.countplot(x='food_category_id', data=df)
ax.set_title('Food Category')
```

```python
A = df['food_category_id'].value_counts()
a_df = pd.DataFrame(A)
a_df.head()
```

```python
fig = px.pie(a_df, values='food_category_id' , names=a_df.index,
             title='Food Category')
fig.show()
```

```python
top_nutrients = df['nutrient_id'].value_counts(ascending=False)[:20]
top_nutrients

fig = px.pie(top_nutrients, values='nutrient_id' , names=top_nutrients.index,
             title='Top 20 Nutrients')
fig.show()

```

```python
top_nutrients = df['nutrient_id'].value_counts().sort_values()[:20]
top_nutrients =pd.DataFrame(top_nutrients)
top_nutrients = top_nutrients.rename(columns={'nutrient_id':'count'})
top_nutrients = top_nutrients.sort_values(by='count', ascending=False)[:20]
top_nutrients
# fig = px.bar(top_nutrients,
#              title='Top 20 Nutrients')
# fig.show()
```

```python
fig = px.bar(top_nutrients, y='count',
             title='Top 20 Nutrients')
fig.show()
```

```python
#define figure size
ax = sns.set(rc={'figure.figsize':(15,8)})
#plot the countplot
ax = sns.countplot(x='count', data=top_nutrients)
ax.set_title('Food Category')
```
